```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, MetaData, Table, select
from datetime import datetime as dt
import sqlite3
import csv
```

```python
yahooData = pd.read_csv("yahoo.csv")
googleData = pd.read_csv("google.csv")
yahooData["Date"] = pd.to_datetime(yahooData["Date"])
googleData["Date"] = pd.to_datetime(googleData["Date"])
yd = yahooData.set_index("Date")
gd = googleData.set_index("Date")
```

```python
# Fiscal Year dates
startDate ='2020-07-01'
endDate = '2021-06-07'
yahooFY = yd.loc[startDate:endDate].reset_index()
googleFY = gd.loc[startDate:endDate].reset_index()


# Quarter1 dates
sdQ1 ='2020-07-01'
edQ1 = '2020-09-30'
yahooQ1 = yd.loc[sdQ1:edQ1].reset_index()
googleQ1 = gd.loc[sdQ1:edQ1].reset_index()


# Quarter2 dates
sdQ2 ='2020-10-01'
edQ2 = '2020-12-31'
yahooQ2 = yd.loc[sdQ2:edQ2].reset_index()
googleQ2 = gd.loc[sdQ2:edQ2].reset_index()


# Quarter3 dates
sdQ3 ='2021-01-01'
edQ3 = '2021-03-31'
yahooQ3 = yd.loc[sdQ3:edQ3].reset_index()
googleQ3 = gd.loc[sdQ3:edQ3].reset_index()


# Quarter4 dates
sdQ4 ='2021-04-01'
edQ4 = '2021-06-07'
yahooQ4 = yd.loc[sdQ4:edQ4].reset_index()
googleQ4 = gd.loc[sdQ4:edQ4].reset_index()

```

```python
# Create Yahoo data dictionary
yahooSY = {}

Q1 = {}

Q1['Dates'] = list(yahooQ1['Date'].astype('string'))
Q1['Highs'] = list(yahooQ1['High'])
Q1['Lows'] = list(yahooQ1['Low'])
Q1['Closes'] = list(yahooQ1['Close'])
Q1['Volumes'] = list(yahooQ1['Volume'])


Q2 = {}

Q2['Dates'] = list(yahooQ2['Date'].astype('string'))
Q2['Highs'] = list(yahooQ2['High'])
Q2['Lows'] = list(yahooQ2['Low'])
Q2['Closes'] = list(yahooQ2['Close'])
Q2['Volumes'] = list(yahooQ2['Volume'])


Q3 = {}

Q3['Dates'] = list(yahooQ3['Date'].astype('string'))
Q3['Highs'] = list(yahooQ3['High'])
Q3['Lows'] = list(yahooQ3['Low'])
Q3['Closes'] = list(yahooQ3['Close'])
Q3['Volumes'] = list(yahooQ3['Volume'])


Q4 = {}

Q4['Dates'] = list(yahooQ4['Date'].astype('string'))
Q4['Highs'] = list(yahooQ4['High'])
Q4['Lows'] = list(yahooQ4['Low'])
Q4['Closes'] = list(yahooQ4['Close'])
Q4['Volumes'] = list(yahooQ4['Volume'])


yahooSY["Quarter1"] = Q1
yahooSY["Quarter2"] = Q2
yahooSY["Quarter3"] = Q3
yahooSY["Quarter4"] = Q4
```

```python
pd.DataFrame(yahooSY).to_json('yahooSD.json')
```
